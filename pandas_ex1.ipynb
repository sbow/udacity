```python
import pandas as pd
import numpy as np
from datetime import datetime
import matplotlib.pyplot as plt
headers = ['Inches'] 
file = 'Book1.csv'
dtypes = [float] 
d = pd.read_csv(file, sep=',', header=0, names=headers, dtype = {"Inches" : np.float})
d.index = pd.to_datetime(d.index) # convert text date mm/dd/yyyy to datetime obj
d.loc['2010-01-01':'2010-01-12'] # get records in between dates (inclusive) using index
dr = d.loc['2010-01-01':'2010-01-12']
dr[(dr > 0.05).all(axis=1)] # data in range greater than certain amount of rain
d['date'] = d.index # Make additional column out of the index (dates)
#d['date'].month==12
dyr = d.resample('M').sum() # sum of rain AS:yearly, W:weekly, M: month, T: minute; more https://stackoverflow.com/questions/17001389/pandas-resample-documentation
#Below: Magic Jupyter line to tell it to plot inline with cell
%matplotlib inline
dyr.plot() # Use pandas plotting function
```
